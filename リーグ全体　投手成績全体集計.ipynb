```python
import pandas as pd
import glob
import os
```

```python
#セリーグ投手成績全チーム分読み込み
files = glob.glob('エクセルファイル/チーム成績/セリーグ/*/投手成績/チーム全体/加工前/投手成績.xlsx')
list = []
for file in files:
    list.append(pd.read_excel(file))
dff = pd.concat(list)
df = dff.loc[:,'試合全体の投球数':'ゲームID']
```

```python
savepath = 'エクセルファイル/チーム成績/リーグ全体/セリーグ全体/投手成績/加工前'
os.makedirs(savepath , exist_ok=True)
pd.DataFrame(df).to_excel(f"{savepath}/全体成績.xlsx", index = False)

print('全体成績終了')

#全体カウント別
df00 = df[(df['B'] == 0)  & (df['S'] == 0)]# 0B 0S
df10 = df[(df['B'] == 1)  & (df['S'] == 0)]# 1B 0S
df20 = df[(df['B'] == 2)  & (df['S'] == 0)]# 2B 0S
df30 = df[(df['B'] == 3)  & (df['S'] == 0)]# 3B 0S
df01 = df[(df['B'] == 0)  & (df['S'] == 1)]# 0B 1S
df11 = df[(df['B'] == 1)  & (df['S'] == 1)]# 1B 1S
df21 = df[(df['B'] == 2)  & (df['S'] == 1)]# 2B 1S
df31 = df[(df['B'] == 3)  & (df['S'] == 1)]# 3B 1S
df02 = df[(df['B'] == 0)  & (df['S'] == 2)]# 0B 2S
df12 = df[(df['B'] == 1)  & (df['S'] == 2)]# 1B 2S
df22 = df[(df['B'] == 2)  & (df['S'] == 2)]# 2B 2S
df32 = df[(df['B'] == 3)  & (df['S'] == 2)]# 3B 2S
df0S = df[df['S'] == 0]# 0S時通算
df1S = df[df['S'] == 1]# 1S時通算
df2S = df[df['S'] == 2]# 2S時通算
df0B = df[df['B'] == 0]# 0B時通算
df1B = df[df['B'] == 1]# 1B時通算
df2B = df[df['B'] == 2]# 2B時通算
df3B = df[df['B'] == 3]# 3B時通算

os.makedirs(f"{savepath}/カウント別" , exist_ok=True)
pd.DataFrame(df00).to_excel(f"{savepath}/カウント別/カウント00.xlsx", index = False)
pd.DataFrame(df10).to_excel(f"{savepath}/カウント別/カウント10.xlsx", index = False)
pd.DataFrame(df20).to_excel(f"{savepath}/カウント別/カウント20.xlsx", index = False)
pd.DataFrame(df30).to_excel(f"{savepath}/カウント別/カウント30.xlsx", index = False)
pd.DataFrame(df01).to_excel(f"{savepath}/カウント別/カウント01.xlsx", index = False)
pd.DataFrame(df11).to_excel(f"{savepath}/カウント別/カウント11.xlsx", index = False)
pd.DataFrame(df21).to_excel(f"{savepath}/カウント別/カウント21.xlsx", index = False)
pd.DataFrame(df31).to_excel(f"{savepath}/カウント別/カウント31.xlsx", index = False)
pd.DataFrame(df02).to_excel(f"{savepath}/カウント別/カウント02.xlsx", index = False)
pd.DataFrame(df12).to_excel(f"{savepath}/カウント別/カウント12.xlsx", index = False)
pd.DataFrame(df22).to_excel(f"{savepath}/カウント別/カウント22.xlsx", index = False)
pd.DataFrame(df32).to_excel(f"{savepath}/カウント別/カウント32.xlsx", index = False)
pd.DataFrame(df0S).to_excel(f"{savepath}/カウント別/カウントs0.xlsx", index = False)
pd.DataFrame(df1S).to_excel(f"{savepath}/カウント別/カウントs1.xlsx", index = False)
pd.DataFrame(df2S).to_excel(f"{savepath}/カウント別/カウントs2.xlsx", index = False)
pd.DataFrame(df0B).to_excel(f"{savepath}/カウント別/カウントb0.xlsx", index = False)
pd.DataFrame(df1B).to_excel(f"{savepath}/カウント別/カウントb1.xlsx", index = False)
pd.DataFrame(df2B).to_excel(f"{savepath}/カウント別/カウントb2.xlsx", index = False)
pd.DataFrame(df3B).to_excel(f"{savepath}/カウント別/カウントb3.xlsx", index = False)

print('カウント終了')

#塁別成績
dfCR = df[(df['2塁'] == 1)  | (df['3塁'] == 1)]    #得点圏打席
df0R = df[(df['1塁'] == 0) & (df['2塁'] == 0) & (df['3塁'] == 0)]    #ランナー無し
df1R = df[(df['1塁'] == 1) & (df['2塁'] == 0) & (df['3塁'] == 0)]    #ランナー１塁
df2R = df[(df['1塁'] == 0) & (df['2塁'] == 1) & (df['3塁'] == 0)]    #ランナー２塁
df3R = df[(df['1塁'] == 0) & (df['2塁'] == 0) & (df['3塁'] == 1)]    #ランナー３塁
df12R = df[(df['1塁'] == 1) & (df['2塁'] == 1) & (df['3塁'] == 0)]    #ランナー１、２塁
df13R = df[(df['1塁'] == 1) & (df['2塁'] == 0) & (df['3塁'] == 1)]    #ランナー１，３塁
df23R = df[(df['1塁'] == 0) & (df['2塁'] == 1) & (df['3塁'] == 1)]    #ランナー２、３塁
df123R = df[(df['1塁'] == 1) & (df['2塁'] == 1) & (df['3塁'] == 1)]   #ランナー満塁

os.makedirs(f"{savepath}/塁別" , exist_ok=True)
pd.DataFrame(df0R).to_excel(f"{savepath}/塁別/R0.xlsx", index = False)
pd.DataFrame(df1R).to_excel(f"{savepath}/塁別/R1.xlsx", index = False)
pd.DataFrame(df2R).to_excel(f"{savepath}/塁別/R2.xlsx", index = False)
pd.DataFrame(df3R).to_excel(f"{savepath}/塁別/R3.xlsx", index = False)
pd.DataFrame(df12R).to_excel(f"{savepath}/塁別/R12.xlsx", index = False)
pd.DataFrame(df13R).to_excel(f"{savepath}/塁別/R13.xlsx", index = False)
pd.DataFrame(df23R).to_excel(f"{savepath}/塁別/R23.xlsx", index = False)
pd.DataFrame(df123R).to_excel(f"{savepath}/塁別/R123.xlsx", index = False)

print('塁別終了')

#得点圏コース別、カウント別
os.makedirs(f"{savepath}/得点圏" , exist_ok=True)
pd.DataFrame(dfCR).to_excel(f"{savepath}/得点圏/得点圏全球成績.xlsx", index = False)

for n in range(1, 26):
    dfC = dfCR[dfCR['コース'] == n]#コース1~25まで
    pd.DataFrame(dfC).to_excel(f"{savepath}/得点圏/{n}.xlsx", index = False)
    
df00 = dfCR[(dfCR['B'] == 0)  & (dfCR['S'] == 0)]# 0B 0S
df10 = dfCR[(dfCR['B'] == 1)  & (dfCR['S'] == 0)]# 1B 0S
df20 = dfCR[(dfCR['B'] == 2)  & (dfCR['S'] == 0)]# 2B 0S
df30 = dfCR[(dfCR['B'] == 3)  & (dfCR['S'] == 0)]# 3B 0S
df01 = dfCR[(dfCR['B'] == 0)  & (dfCR['S'] == 1)]# 0B 1S
df11 = dfCR[(dfCR['B'] == 1)  & (dfCR['S'] == 1)]# 1B 1S
df21 = dfCR[(dfCR['B'] == 2)  & (dfCR['S'] == 1)]# 2B 1S
df31 = dfCR[(dfCR['B'] == 3)  & (dfCR['S'] == 1)]# 3B 1S
df02 = dfCR[(dfCR['B'] == 0)  & (dfCR['S'] == 2)]# 0B 2S
df12 = dfCR[(dfCR['B'] == 1)  & (dfCR['S'] == 2)]# 1B 2S
df22 = dfCR[(dfCR['B'] == 2)  & (dfCR['S'] == 2)]# 2B 2S
df32 = dfCR[(dfCR['B'] == 3)  & (dfCR['S'] == 2)]# 3B 2S
df0S = dfCR[dfCR['S'] == 0]# 0S時通算
df1S = dfCR[dfCR['S'] == 1]# 1S時通算
df2S = dfCR[dfCR['S'] == 2]# 2S時通算
df0B = dfCR[dfCR['B'] == 0]# 0B時通算

df1B = dfCR[dfCR['B'] == 1]# 1B時通算
df2B = dfCR[dfCR['B'] == 2]# 2B時通算
df3B = dfCR[dfCR['B'] == 3]# 3B時通算

pd.DataFrame(df00).to_excel(f"{savepath}/得点圏/カウント00.xlsx", index = False)
pd.DataFrame(df10).to_excel(f"{savepath}/得点圏/カウント10.xlsx", index = False)
pd.DataFrame(df20).to_excel(f"{savepath}/得点圏/カウント20.xlsx", index = False)
pd.DataFrame(df30).to_excel(f"{savepath}/得点圏/カウント30.xlsx", index = False)
pd.DataFrame(df01).to_excel(f"{savepath}/得点圏/カウント01.xlsx", index = False)
pd.DataFrame(df11).to_excel(f"{savepath}/得点圏/カウント11.xlsx", index = False)
pd.DataFrame(df21).to_excel(f"{savepath}/得点圏/カウント21.xlsx", index = False)
pd.DataFrame(df31).to_excel(f"{savepath}/得点圏/カウント31.xlsx", index = False)
pd.DataFrame(df02).to_excel(f"{savepath}/得点圏/カウント02.xlsx", index = False)
pd.DataFrame(df12).to_excel(f"{savepath}/得点圏/カウント12.xlsx", index = False)
pd.DataFrame(df22).to_excel(f"{savepath}/得点圏/カウント22.xlsx", index = False)
pd.DataFrame(df32).to_excel(f"{savepath}/得点圏/カウント32.xlsx", index = False)
pd.DataFrame(df0S).to_excel(f"{savepath}/得点圏/カウントs0.xlsx", index = False)
pd.DataFrame(df1S).to_excel(f"{savepath}/得点圏/カウントs1.xlsx", index = False)
pd.DataFrame(df2S).to_excel(f"{savepath}/得点圏/カウントs2.xlsx", index = False)
pd.DataFrame(df0B).to_excel(f"{savepath}/得点圏/カウントb0.xlsx", index = False)
pd.DataFrame(df1B).to_excel(f"{savepath}/得点圏/カウントb1.xlsx", index = False)
pd.DataFrame(df2B).to_excel(f"{savepath}/得点圏/カウントb2.xlsx", index = False)
pd.DataFrame(df3B).to_excel(f"{savepath}/得点圏/カウントb3.xlsx", index = False)
    
print('得点圏終了')

    
#コース別
os.makedirs(f"{savepath}/コース別" , exist_ok=True)
for n in range(1, 26):
    dfC = df[df['コース'] == n]#コース1~25まで
    pd.DataFrame(dfC).to_excel(f"{savepath}/コース別/{n}.xlsx", index = False)
    
print('コース終了')
    

#球場別
Stadiums = df['球場'].unique()
os.makedirs(f"{savepath}/球場別" , exist_ok=True)
for n in range(0, len(Stadiums)):
    Stadium = Stadiums[n]
    dfS = df[df['球場'] == Stadium]
    pd.DataFrame(dfS).to_excel(f"{savepath}/球場別/{Stadium}.xlsx", index = False)
    
print('球場別終了')
    
    
#対右
os.makedirs(f"{savepath}/対左右/右" , exist_ok=True)
dfPHR = df[df['左右打'] == '右打']
pd.DataFrame(dfPHR).to_excel(f"{savepath}/対左右/右/全体成績.xlsx", index = False)
for n in range(1, 26):
    dfC = dfPHR[dfPHR['コース'] == n]#コース1~25まで
    pd.DataFrame(dfC).to_excel(f"{savepath}/対左右/右/{n}.xlsx", index = False)
    
#対左
os.makedirs(f"{savepath}/対左右/左" , exist_ok=True)
dfPHL = df[df['左右打'] == '左打']
pd.DataFrame(dfPHL).to_excel(f"{savepath}/対左右/左/全体成績.xlsx", index = False)
for n in range(1, 26):
    dfC = dfPHL[dfPHL['コース'] == n]#コース1~25まで
    pd.DataFrame(dfC).to_excel(f"{savepath}/対左右/左/{n}.xlsx", index = False)
    
print('対左右終了')    

#点差別成績
df0 = df[df['点差'] == 0]# 同点時
df1_3 = df[(df['点差'] >= 1)  & (df['点差'] <= 3)]# 1~3点リード時
df3_ = df[df['点差'] >= 4]# 4点以上リード時
dfB1_3 = df[(df['点差'] <= -1)  & (df['点差'] >= -3)]# 1~3点ビハインド時
dfB3_ = df[df['点差'] <= -4]# 4点以上ビハインド時
    
#逆転シチュエーション
df0P = dfCR[dfCR['点差'] == 0]#同点時、得点圏
df1P = dfCR[dfCR['点差'] == -1]
df2P = dfCR[dfCR['点差'] == -2]
df3P = df123R[df123R['点差'] == -3]#3点差、満塁

os.makedirs(f"{savepath}/点差別" , exist_ok=True)
pd.DataFrame(df0).to_excel(f"{savepath}/点差別/同点時.xlsx", index = False)
pd.DataFrame(df1_3).to_excel(f"{savepath}/点差別/1~3点リード時.xlsx", index = False)
pd.DataFrame(df3_).to_excel(f"{savepath}/点差別/4点以上リード時.xlsx", index = False)
pd.DataFrame(dfB1_3).to_excel(f"{savepath}/点差別/1~3点ビハインド時.xlsx", index = False)
pd.DataFrame(dfB3_).to_excel(f"{savepath}/点差別/4点以上ビハインド時.xlsx", index = False)
pd.DataFrame(df0P).to_excel(f"{savepath}/点差別/同点時得点圏.xlsx", index = False)
pd.DataFrame(df1P).to_excel(f"{savepath}/点差別/1点差得点圏.xlsx", index = False)
pd.DataFrame(df2P).to_excel(f"{savepath}/点差別/2点差得点圏.xlsx", index = False)
pd.DataFrame(df3P).to_excel(f"{savepath}/点差別/3点差満塁時.xlsx", index = False)

#曜日別成績
weeks = ['（月）','（火）','（水）','（木）','（金）','（土）','（日）']
os.makedirs(f"{savepath}/曜日別" , exist_ok=True)
for w in range(0,len(weeks)):
    week = weeks[w]
    dfw = df[df['試合日'].str.contains(week)]
    pd.DataFrame(dfw).to_excel(f"{savepath}/曜日別/{week}.xlsx", index = False)
    
#月別成績
months = ['3月','4月','5月','6月','7月','8月','9月','10月','11月',]
os.makedirs(f"{savepath}/月別" , exist_ok=True)
for m in range(0,len(months)):
    month = months[m]
    dfm = df[df['試合日'].str.contains(month)]
    pd.DataFrame(dfm).to_excel(f"{savepath}/月別/{month}.xlsx", index = False)
    
#チーム内打順別成績
Bat_O = df['打順'].unique()
os.makedirs(f"{savepath}/打順別" , exist_ok=True)
for o in range(0, len(Bat_O)):
    Bat = Bat_O[o]
    dfBO = df[df['打順'] == Bat]
    pd.DataFrame(dfBO).to_excel(f"{savepath}/打順別/{Bat}.xlsx", index = False)
        
#チーム内守備位置別成績
positions = ['投', '捕', '一', '二', '三', '遊', '左', '中', '右', '指', '打']
os.makedirs(f"{savepath}/守備位置別" , exist_ok=True)
for p in range(0, len(positions)):
    #選手ごと全体成績
    position = positions[p]
    dfBO = df[df['打者守備位置'] == position]
    pd.DataFrame(dfBO).to_excel(f"{savepath}/守備位置別/{position}.xlsx", index = False)
    
#イニング別成績
Innings = ['1回', '2回', '3回', '4回', '5回', '6回', '7回', '8回', '9回', ]
os.makedirs(f"{savepath}/イニング別" , exist_ok=True)
for ini in range(0, len(Innings)):
    Inning = Innings[ini]
    dfIni = df[(df['何回'].str.contains(Inning)) & (~df['何回'].str.contains(f"\d{Inning}"))]
    pd.DataFrame(dfIni).to_excel(f"{savepath}/イニング別/{Inning}.xlsx", index = False)
        
dfI_O = df[df['何回'].str.len() >= 4]
pd.DataFrame(dfI_O).to_excel(f"{savepath}/イニング別/延長戦.xlsx", index = False)

print('操作完了')
```
