```python
import pandas as pd
import glob
import os
import openpyxl
from pathlib import Path
```

```python
#打者成績
file_loadpath = 'エクセルファイル/チーム成績/*/*/打者成績/チーム全体/加工前/'
file_savepath1 = 'エクセルファイル/チーム成績'
file_savepath2 = '打者成績/チーム全体/加工後'
Template = 'エクセルファイル/野手 ひな形/'
max_column = 38

#全球成績
files = glob.glob(file_loadpath + '*.xlsx')
for i in range(0, len(files)):
    #ひな形読み込み
    wb = openpyxl.load_workbook(f"{Template}全球.xlsx")
    ws1 = wb.worksheets[0]
    #コピーしたいもの読み込み
    file = files[i]
    wbs = openpyxl.load_workbook(file).worksheets[0]
    #列はゲームIDまで、行は入力されてる行の最大値までとる
    max_row = wbs.max_row

    #ひな形へコピーしていく
    for r in range(1, max_row+1):
            for c in range(1, max_column+1):
                 # セル内容のコピー
                 ws1.cell(row = r,column = c).value = wbs.cell(row = r,column = c).value
            
    Data_league = Path(file).parts[-6]
    Data_team = Path(file).parts[-5]
    Data_dirs = f"{file_savepath1}/{Data_league}/{Data_team}/{file_savepath2}"
    os.makedirs(Data_dirs , exist_ok=True)
    #保存
    wb.save(f"{Data_dirs}/全球成績 {Data_team}.xlsx")

print('全球完了')
    
#塁別まとめ
files_Base = glob.glob(file_loadpath + '塁別/*.xlsx')
roop = int(len(files_Base) / 8) #8ファイルで1周
for num in range(0,roop): #全周文繰り返し
    roop_T = num * 8
    wb_new = openpyxl.load_workbook(f"{Template}塁別ひな形.xlsx")
    for i in range(0,8): #8ファイルで1周分繰り返し
        num2 = roop_T + i
        #コピーしたいもの読み込み
        file_Base = files_Base[num2]
        wbs = openpyxl.load_workbook(file_Base).worksheets[0]
        basename_without_ext = os.path.splitext(os.path.basename(file_Base))[0]#ファイル名取得
        #コピー先セル指定
        ws1 = wb_new[basename_without_ext]
        #列はゲームIDまで、行は入力されてる行の最大値までとる
        max_row = wbs.max_row
        
        for r in range(2, max_row+1):
            for c in range(1, max_column+1):
                # セル内容のコピー
                ws1.cell(row = r+29 ,column = c).value = wbs.cell(row = r,column = c).value

    Data_league = Path(file_Base).parts[-7]
    Data_team = Path(file_Base).parts[-6]
    Data_dirs = f"{file_savepath1}/{Data_league}/{Data_team}/{file_savepath2}"
    wb_new.save(f"{Data_dirs}/塁別成績 {Data_team}.xlsx")
    
print('塁別終了')
    
#カウント別まとめ
files_Count = glob.glob(file_loadpath + 'カウント別/*.xlsx')
roop = int(len(files_Count) / 19) #19ファイルで1周

for num in range(0,roop): #全周文繰り返し
    roop_T = num * 19
    wb_new = openpyxl.load_workbook(f"{Template}カウント別ひな形.xlsx")
    for i in range(0,19): #19ファイルで1周分繰り返し
        num2 = roop_T + i
        #コピーしたいもの読み込み
        file_Count = files_Count[num2]
        wbs = openpyxl.load_workbook(file_Count).worksheets[0]
        basename_without_ext = os.path.splitext(os.path.basename(file_Count))[0]#ファイル名取得
        #コピー先セル指定
        ws1 = wb_new[basename_without_ext]
        #列はゲームIDまで、行は入力されてる行の最大値までとる
        max_row = wbs.max_row
        
        for r in range(2, max_row+1):
            for c in range(1, max_column+1):
                # セル内容のコピー
                ws1.cell(row = r+29 ,column = c).value = wbs.cell(row = r,column = c).value

    Data_league = Path(file_Count).parts[-7]
    Data_team = Path(file_Count).parts[-6]
    Data_dirs = f"{file_savepath1}/{Data_league}/{Data_team}/{file_savepath2}"
    wb_new.save(f"{Data_dirs}/カウント別成績 {Data_team}.xlsx")

print('カウント別完了')

#コース別まとめ
files_Course = glob.glob(file_loadpath + 'コース別/*.xlsx')
roop = int(len(files_Course) / 25) #25ファイルで1周

for num in range(0,roop): #全周文繰り返し
    roop_T = num * 25
    wb_new = openpyxl.load_workbook(f"{Template}コース別ひな形.xlsx")
    for i in range(0,25): #25ファイルで1周分繰り返し
        num2 = roop_T + i
        #コピーしたいもの読み込み
        file_Course = files_Course[num2]
        wbs = openpyxl.load_workbook(file_Course).worksheets[0]
        basename_without_ext = os.path.splitext(os.path.basename(file_Course))[0]#ファイル名取得
        #コピー先セル指定
        ws1 = wb_new[basename_without_ext]
        #列はゲームIDまで、行は入力されてる行の最大値までとる
        max_row = wbs.max_row
        
        for r in range(2, max_row+1):
            for c in range(1, max_column+1):
                # セル内容のコピー
                ws1.cell(row = r+29 ,column = c).value = wbs.cell(row = r,column = c).value
    
    Data_league = Path(file_Course).parts[-7]
    Data_team = Path(file_Course).parts[-6]
    Data_dirs = f"{file_savepath1}/{Data_league}/{Data_team}/{file_savepath2}"
    wb_new.save(f"{Data_dirs}/コース別成績 {Data_team}.xlsx")
    
print('コース別完了')
    
#対右
files_R = glob.glob(file_loadpath + '対左右/右/*.xlsx')
roop = int(len(files_R) / 26) #26ファイルで1周

for num in range(0,roop): #全周文繰り返し
    roop_T = num * 26
    wb_new = openpyxl.load_workbook(f"{Template}対右ひな形.xlsx")
    for i in range(0,26): #26ファイルで1周分繰り返し
        num2 = roop_T + i
        #コピーしたいもの読み込み
        file_R = files_R[num2]
        wbs = openpyxl.load_workbook(file_R).worksheets[0]
        basename_without_ext = os.path.splitext(os.path.basename(file_R))[0]#ファイル名取得
        #コピー先セル指定
        ws1 = wb_new[basename_without_ext]
        #列はゲームIDまで、行は入力されてる行の最大値までとる
        max_row = wbs.max_row
        
        for r in range(2, max_row+1):
            for c in range(1, max_column+1):
                # セル内容のコピー
                ws1.cell(row = r+29 ,column = c).value = wbs.cell(row = r,column = c).value
    
    Data_league = Path(file_R).parts[-8]
    Data_team = Path(file_R).parts[-7]
    Data_dirs = f"{file_savepath1}/{Data_league}/{Data_team}/{file_savepath2}"
    wb_new.save(f"{Data_dirs}/対右成績 {Data_team}.xlsx")

print('対右完了')
    
#対左
files_L = glob.glob(file_loadpath + '対左右/左/*.xlsx')
roop = int(len(files_L) / 26) #26ファイルで1周

for num in range(0,roop): #全周文繰り返し
    roop_T = num * 26
    wb_new = openpyxl.load_workbook(f"{Template}対左ひな形.xlsx")
    for i in range(0,26): #26ファイルで1周分繰り返し
        num2 = roop_T + i
        #コピーしたいもの読み込み
        file_L = files_L[num2]
        wbs = openpyxl.load_workbook(file_L).worksheets[0]
        basename_without_ext = os.path.splitext(os.path.basename(file_L))[0]#ファイル名取得
        #コピー先セル指定
        ws1 = wb_new[basename_without_ext]
        #列はゲームIDまで、行は入力されてる行の最大値までとる
        max_row = wbs.max_row
        
        for r in range(2, max_row+1):
            for c in range(1, max_column+1):
                # セル内容のコピー
                ws1.cell(row = r+29 ,column = c).value = wbs.cell(row = r,column = c).value
    
    Data_league = Path(file_L).parts[-8]
    Data_team = Path(file_L).parts[-7]
    Data_dirs = f"{file_savepath1}/{Data_league}/{Data_team}/{file_savepath2}"
    wb_new.save(f"{Data_dirs}/対左成績 {Data_team}.xlsx")

print('対左完了')

#球場別成績まとめ
files_Stadium = glob.glob(file_loadpath + '球場別/*.xlsx')
wb_new = openpyxl.load_workbook(f"{Template}球場別ひな形.xlsx")
n = 0
for i in range(0,len(files_Stadium)):
    n += 1
    #コピーしたいもの読み込み
    file_Stadium = files_Stadium[i]
    wbs = openpyxl.load_workbook(file_Stadium).worksheets[0]
    basename_without_ext = os.path.splitext(os.path.basename(file_Stadium))[0]#ファイル名取得
    #コピー先セル指定
    ws1 = wb_new.worksheets[n-1]
    ws1.title = basename_without_ext
    #列はゲームIDまで、行は入力されてる行の最大値までとる
    max_row = wbs.max_row
        
    for r in range(2, max_row+1):
        for c in range(1, max_column+1):
            # セル内容のコピー
            ws1.cell(row = r+29 ,column = c).value = wbs.cell(row = r,column = c).value

    #現在操作中ファイルの打者名
    Data_team = Path(file_Stadium).parts[-6]
    
    #最後のファイルの時、保存して終わる
    if i+1 == len(files_Stadium):
        Data_league = Path(file_Stadium).parts[-7]
        Data_dirs = f"{file_savepath1}/{Data_league}/{Data_team}/{file_savepath2}"
        for m in range(n,30):
                wb_new.remove(wb_new.worksheets[n])
        wb_new.save(f"{Data_dirs}/球場別成績 {Data_team}.xlsx")
        break

    else:
        #次操作予定の打者名抜き出し
        file_Stadium_Next = files_Stadium[i+1]
        Data_team_Next = Path(file_Stadium_Next).parts[-6]

        if Data_team != Data_team_Next:
            Data_league = Path(file_Stadium).parts[-7]
            Data_dirs = f"{file_savepath1}/{Data_league}/{Data_team}/{file_savepath2}"
            for m in range(n,30):
                wb_new.remove(wb_new.worksheets[n])
            wb_new.save(f"{Data_dirs}/球場別成績 {Data_team}.xlsx")
            wb_new = openpyxl.load_workbook(f"{Template}球場別ひな形.xlsx")
            n = 0

print('球場別完了')

#対チーム別まとめ
files_Team = glob.glob(file_loadpath + '相手チーム別/*.xlsx')
roop = int(len(files_Team) / 12) #12ファイルで1周

for num in range(0,roop): #全周文繰り返し
    roop_T = num * 12
    wb_new = openpyxl.load_workbook(f"{Template}相手チーム別ひな形.xlsx")
    for i in range(0,12): #12ファイルで1周分繰り返し
        num2 = roop_T + i
        #コピーしたいもの読み込み
        file_Team = files_Team[num2]
        wbs = openpyxl.load_workbook(file_Team).worksheets[0]
        basename_without_ext = os.path.splitext(os.path.basename(file_Team))[0]#ファイル名取得
        #コピー先セル指定
        ws1 = wb_new[basename_without_ext]
        #列はゲームIDまで、行は入力されてる行の最大値までとる
        max_row = wbs.max_row
        
        for r in range(1, max_row+1):
            for c in range(1, max_column+1):
                # セル内容のコピー
                ws1.cell(row = r+29 ,column = c).value = wbs.cell(row = r,column = c).value

    Data_league = Path(file_Team).parts[-7]
    Data_team = Path(file_Team).parts[-6]
    Data_dirs = f"{file_savepath1}/{Data_league}/{Data_team}/{file_savepath2}"
    wb_new.save(f"{Data_dirs}/相手チーム別成績 {Data_team}.xlsx")

print('対チーム別完了')    
    
#得点圏まとめ
files_C = glob.glob(file_loadpath + '得点圏/*.xlsx')
roop = int(len(files_C) / 45) #45ファイルで1周

for num in range(0,roop): #全周文繰り返し
    roop_T = num * 45
    wb_new = openpyxl.load_workbook(f"{Template}得点圏ひな形.xlsx")
    for i in range(0,45): #45ファイルで1周分繰り返し
        num2 = roop_T + i
        #コピーしたいもの読み込み
        file_C = files_C[num2]
        wbs = openpyxl.load_workbook(file_C).worksheets[0]
        basename_without_ext = os.path.splitext(os.path.basename(file_C))[0]#ファイル名取得
        #コピー先セル指定
        ws1 = wb_new[basename_without_ext]
        #列はゲームIDまで、行は入力されてる行の最大値までとる
        max_row = wbs.max_row
        
        for r in range(2, max_row+1):
            for c in range(1, max_column+1):
                # セル内容のコピー
                ws1.cell(row = r+29 ,column = c).value = wbs.cell(row = r,column = c).value

    Data_league = Path(file_C).parts[-7]
    Data_team = Path(file_C).parts[-6]
    Data_dirs = f"{file_savepath1}/{Data_league}/{Data_team}/{file_savepath2}"
    wb_new.save(f"{Data_dirs}/得点圏成績 {Data_team}.xlsx")
    
print('得点圏完了')
    
#点差別まとめ
files_PD = glob.glob(file_loadpath + '点差別/*.xlsx')
roop = int(len(files_PD) / 9) #9ファイルで1周

for num in range(0,roop): #全周文繰り返し
    roop_T = num * 9
    wb_new = openpyxl.load_workbook(f"{Template}点差別ひな形.xlsx")
    for i in range(0,9): #9ファイルで1周分繰り返し
        num2 = roop_T + i
        #コピーしたいもの読み込み
        file_PD = files_PD[num2]
        wbs = openpyxl.load_workbook(file_PD).worksheets[0]
        basename_without_ext = os.path.splitext(os.path.basename(file_PD))[0]#ファイル名取得
        #コピー先セル指定
        ws1 = wb_new[basename_without_ext]
        #列はゲームIDまで、行は入力されてる行の最大値までとる
        max_row = wbs.max_row
        
        for r in range(1, max_row+1):
            for c in range(1, max_column+1):
                # セル内容のコピー
                ws1.cell(row = r+29 ,column = c).value = wbs.cell(row = r,column = c).value

    Data_league = Path(file_PD).parts[-7]
    Data_team = Path(file_PD).parts[-6]
    Data_dirs = f"{file_savepath1}/{Data_league}/{Data_team}/{file_savepath2}"
    wb_new.save(f"{Data_dirs}/点差別成績 {Data_team}.xlsx")
    
print('点差別完了')


#曜日別まとめ
files_week = glob.glob(file_loadpath + '曜日別/*.xlsx')
roop = int(len(files_week) / 7) #7ファイルで1周

for num in range(0,roop): #全周文繰り返し
    roop_T = num * 7
    wb_new = openpyxl.load_workbook(f"{Template}曜日別ひな形.xlsx")
    for i in range(0,7): #9ファイルで1周分繰り返し
        num2 = roop_T + i
        #コピーしたいもの読み込み
        file_week = files_week[num2]
        wbs = openpyxl.load_workbook(file_week).worksheets[0]
        basename_without_ext = os.path.splitext(os.path.basename(file_week))[0]#ファイル名取得
        #コピー先セル指定
        ws1 = wb_new[basename_without_ext]
        #列はゲームIDまで、行は入力されてる行の最大値までとる
        max_row = wbs.max_row
        
        for r in range(1, max_row+1):
            for c in range(1, max_column+1):
                # セル内容のコピー
                ws1.cell(row = r+29 ,column = c).value = wbs.cell(row = r,column = c).value

    Data_league = Path(file_week).parts[-7]
    Data_team = Path(file_week).parts[-6]
    Data_dirs = f"{file_savepath1}/{Data_league}/{Data_team}/{file_savepath2}"
    wb_new.save(f"{Data_dirs}/曜日別成績 {Data_team}.xlsx")
    
print('曜日別完了')

#月別まとめ
files_month = glob.glob(file_loadpath + '月別/*.xlsx')
roop = int(len(files_month) / 9) #9ファイルで1周

for num in range(0,roop): #全周文繰り返し
    roop_T = num * 9
    wb_new = openpyxl.load_workbook(f"{Template}月別ひな形.xlsx")
    for i in range(0,9): #9ファイルで1周分繰り返し
        num2 = roop_T + i
        #コピーしたいもの読み込み
        file_month = files_month[num2]
        wbs = openpyxl.load_workbook(file_month).worksheets[0]
        basename_without_ext = os.path.splitext(os.path.basename(file_month))[0]#ファイル名取得
        #コピー先セル指定
        ws1 = wb_new[basename_without_ext]
        #列はゲームIDまで、行は入力されてる行の最大値までとる
        max_row = wbs.max_row
        
        for r in range(1, max_row+1):
            for c in range(1, max_column+1):
                # セル内容のコピー
                ws1.cell(row = r+29 ,column = c).value = wbs.cell(row = r,column = c).value

    Data_league = Path(file_month).parts[-7]
    Data_team = Path(file_month).parts[-6]
    Data_dirs = f"{file_savepath1}/{Data_league}/{Data_team}/{file_savepath2}"
    wb_new.save(f"{Data_dirs}/月別成績 {Data_team}.xlsx")
    
print('月別完了')

    
#打順別まとめ
files_Order = glob.glob(file_loadpath + '打順別/*.xlsx')
roop = int(len(files_Order) / 9) #9ファイルで1周

for num in range(0,roop): #全周文繰り返し
    roop_T = num * 9
    wb_new = openpyxl.load_workbook(f"{Template}打順別ひな形.xlsx")
    for i in range(0,9): #9ファイルで1周分繰り返し
        num2 = roop_T + i
        #コピーしたいもの読み込み
        file_Order = files_Order[num2]
        wbs = openpyxl.load_workbook(file_Order).worksheets[0]
        basename_without_ext = os.path.splitext(os.path.basename(file_Order))[0]#ファイル名取得
        #コピー先セル指定
        ws1 = wb_new[basename_without_ext]
        #列はゲームIDまで、行は入力されてる行の最大値までとる
        max_row = wbs.max_row
        
        for r in range(1, max_row+1):
            for c in range(1, max_column+1):
                # セル内容のコピー
                ws1.cell(row = r+29 ,column = c).value = wbs.cell(row = r,column = c).value

    Data_league = Path(file_Order).parts[-7]
    Data_team = Path(file_Order).parts[-6]
    Data_dirs = f"{file_savepath1}/{Data_league}/{Data_team}/{file_savepath2}"
    wb_new.save(f"{Data_dirs}/打順別成績 {Data_team}.xlsx")
    
print('打順別完了')

#守備位置別まとめ
files_posi = glob.glob(file_loadpath + '守備位置別/*.xlsx')
roop = int(len(files_posi) / 11) #11ファイルで1周

for num in range(0,roop): #全周文繰り返し
    roop_T = num * 11
    wb_new = openpyxl.load_workbook(f"{Template}守備位置別ひな形.xlsx")
    for i in range(0,11): #11ファイルで1周分繰り返し
        num2 = roop_T + i
        #コピーしたいもの読み込み
        file_posi = files_posi[num2]
        wbs = openpyxl.load_workbook(file_posi).worksheets[0]
        basename_without_ext = os.path.splitext(os.path.basename(file_posi))[0]#ファイル名取得
        #コピー先セル指定
        ws1 = wb_new[basename_without_ext]
        #列はゲームIDまで、行は入力されてる行の最大値までとる
        max_row = wbs.max_row
        
        for r in range(1, max_row+1):
            for c in range(1, max_column+1):
                # セル内容のコピー
                ws1.cell(row = r+29 ,column = c).value = wbs.cell(row = r,column = c).value

    Data_league = Path(file_posi).parts[-7]
    Data_team = Path(file_posi).parts[-6]
    Data_dirs = f"{file_savepath1}/{Data_league}/{Data_team}/{file_savepath2}"
    wb_new.save(f"{Data_dirs}/守備位置別成績 {Data_team}.xlsx")
    
print('守備位置別完了')

#イニング別まとめ
files_Inning = glob.glob(file_loadpath + 'イニング別/*.xlsx')
roop = int(len(files_Inning) / 10) #10ファイルで1周

for num in range(0,roop): #全周文繰り返し
    roop_T = num * 10
    wb_new = openpyxl.load_workbook(f"{Template}イニング別ひな形.xlsx")
    for i in range(0,10): #10ファイルで1周分繰り返し
        num2 = roop_T + i
        #コピーしたいもの読み込み
        file_Inning = files_Inning[num2]
        wbs = openpyxl.load_workbook(file_Inning).worksheets[0]
        basename_without_ext = os.path.splitext(os.path.basename(file_Inning))[0]#ファイル名取得
        #コピー先セル指定
        ws1 = wb_new[basename_without_ext]
        #列はゲームIDまで、行は入力されてる行の最大値までとる
        max_row = wbs.max_row
        
        for r in range(1, max_row+1):
            for c in range(1, max_column+1):
                # セル内容のコピー
                ws1.cell(row = r+29 ,column = c).value = wbs.cell(row = r,column = c).value

    Data_league = Path(file_Inning).parts[-7]
    Data_team = Path(file_Inning).parts[-6]
    Data_dirs = f"{file_savepath1}/{Data_league}/{Data_team}/{file_savepath2}"
    wb_new.save(f"{Data_dirs}/イニング別成績 {Data_team}.xlsx")
    
print('イニング別完了,全終了')
```
