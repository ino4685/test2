```python
import matplotlib.pyplot as plt
import matplotlib.dates as dt
import glob
import os
import pandas as pd
import japanize_matplotlib
from matplotlib import patches
import seaborn as sns
from PIL import Image
%matplotlib inline
import json #標準のjsonモジュールの読み込み
from requests_oauthlib import OAuth1Session #OAuthのライブラリの読み込み
```

```python
#API_Key(CONSUMER_KEY), API_Key_Secret(CONSUMER_SECRET), ACCESS_TOKEN, ACCESS_TOKEN_SECRET
APIK = ''
APIKS = ''
AT = ''
ATS = ''

url_media = "https://upload.twitter.com/1.1/media/upload.json"
url_text = "https://api.twitter.com/1.1/statuses/update.json"

#定数類
guide_c = 'grey'
strike_c = 'r'
strike_z_c = 'silver'
ball_c = [['見逃し0','darkblue'],['空振り1','yellow'],['ファウル2','orangered'],['ゴロアウト3','brown'],['フライアウト4','gray'],['ライナーアウト5','mediumpurple'],
          ['見逃し三振6','dodgerblue'],['空振り三振7','greenyellow'],['併殺8','black'],['犠打9','green'],['犠飛10','violet'],['単打11','aqua'],['長打12','red'],]

Team = '西武'
file_savepath = 'エクセルファイル/全試合元データ/図用'
file_loadpath = 'エクセルファイル/全試合元データ/図用/*.xlsx'
```

```python
files = glob.glob(file_loadpath)
for bb in range(0, len(files)):
    file = files[bb]
    dfSS = pd.read_excel(file)
    df_B = dfSS[(dfSS['後攻（ホーム）'] == Team) & (dfSS['何回'].str.contains('表')) | (dfSS['先攻（アウェイ）'] == Team) & (dfSS['何回'].str.contains('裏'))]
    
    members_B = df_B['投手名'].unique()
    print(f"{len(members_B)}人")
    
    for mem in range(0, len(members_B)):
        #選手ごと全体成績
        member = members_B[mem]
        _df = df_B[df_B['投手名'] == member]
        _df.reset_index(drop=True, inplace=True)
        dfSS = _df.loc[:,['投手名','左右投','打者名','左右打','球種','球速','コース内外','コース高低','試合日','コンタクト','打球種類']]
        P_name = dfSS.iat[0,0]#投手名
        Day = dfSS.iat[0,8]#試合日
        
        #結果関係
        dfR = _df.loc[:,'打球種類']
        sh = (dfR == 7).sum()#単打
        wh = (dfR == 8).sum()#二塁打
        th = (dfR == 9).sum()#三塁打
        Hh = (dfR == 10).sum()#本塁打

        K = (dfR == 6).sum()#三振
        fb = (dfR == 11).sum()#四球
        db = (dfR == 12).sum()#死球
        
        dfRBI = _df.loc[:,'打点']
        po = dfRBI.sum()#失点
        
        Bat_c = (dfR != 0).sum()#打者数
        
        #球種別割合表示
        fig = plt.figure(figsize=(15, 15))
        fig.patch.set_facecolor('white')

        #対右投
        df_R = dfSS[dfSS['左右打'] == '右打']
        P_types = df_R['球種'].value_counts().to_dict()
        label = list(P_types.keys()) #球種
        value = list(P_types.values()) #球種毎投球数

        ax = fig.add_subplot(121,title="対右打者球種別投球割合")
        ax.pie(x=value, autopct='%1.1f%%', startangle=90, counterclock=False,textprops={'fontsize': 15},pctdistance=0.75)
        ax.legend(label,bbox_to_anchor=(1,1), loc='upper left')
        ax.text(1, -1, f"投球数:{sum(value)}")

        #対左投
        df_L = dfSS[dfSS['左右打'] == '左打']
        P_types = df_L['球種'].value_counts().to_dict()
        label = list(P_types.keys()) #球種
        value = list(P_types.values()) #球種毎投球数

        ax = fig.add_subplot(122,title="対左打者球種別投球割合")
        ax.pie(x=value, autopct='%1.1f%%', startangle=90, counterclock=False,textprops={'fontsize': 15},pctdistance=0.75)
        ax.legend(label,bbox_to_anchor=(1,1), loc='upper left')
        ax.text(1, -1, f"投球数:{sum(value)}")
        
        os.makedirs(f"画像/{P_name}/球種割合" , exist_ok=True)
        fig.savefig(f"画像/{P_name}/球種割合/{P_name} {Day}.jpg")
        
        
        #球種別投球チャート表示

        P_types = dfSS['球種'].unique()
        row = len(P_types)

        fig = plt.figure(figsize=(5*(row+1), 10))
        fig.patch.set_facecolor('white')

        image = Image.open("画像/凡例.png")
        ax = fig.add_subplot(f'2{row+1}{row+1}').imshow(image)
        plt.xticks([])
        plt.yticks([])
        
        for ball in range(row):
            P_type = P_types[ball]
            df_P = df_R[df_R['球種'] == P_type]
            ax = fig.add_subplot(2,f'{row+1}',f'{ball+1}',title=f"{P_type} 対右")
            ax.title.set_size(15)
            #x軸設定
            ax.set_xlim(0, 160)
            ax.set_xticks([])
            x_v = [32,64,96,128]
            for i in range(len(x_v)):
                ax.axvline(x_v[i], ls = "-.", color = guide_c)
                if i == 0 or i == 3:
                    ax.axvline(x_v[i], color = strike_c,ymin=0.2, ymax=0.8)

            #y軸設定
            ax.set_yticks([])
            ax.set_ylim(200, 0)
            y_h = [40,80,120,160]
            for j in range(len(y_h)):
                ax.axhline(y_h[j], ls = "-.", color = guide_c)
                if j == 0 or j == 3:
                    ax.axhline(y_h[j], color = strike_c,xmin=0.2, xmax=0.8)

            #ストライクゾーン内の色設定
            r = patches.Rectangle( (32,40) , 96, 120, facecolor=strike_z_c,alpha=0.3)
            ax.add_patch(r)


            for ball2 in range(len(df_P)):
                x = df_P.iat[ball2,6]
                y = df_P.iat[ball2,7]
                co = df_P.iat[ball2,9]
                kind = df_P.iat[ball2,10]
                if kind == 1:
                    c_no = 3
                elif kind == 2:
                    c_no = 4
                elif kind == 3:
                    c_no = 5
                elif kind == 4:
                    c_no = 9
                elif kind == 5:
                    c_no = 10
                elif kind == 6:
                    if co != 0:
                        c_no = 7
                    else:
                        c_no = 6
                elif kind == 7:
                    c_no = 11
                elif kind == 8 or kind == 9 or kind == 10:
                    c_no = 12
                elif kind == 14:
                    c_no = 8
                elif kind == 0:
                    if co == 2:
                        c_no = 2
                    elif co == 1:
                        c_no = 1
                    else:
                        c_no = 0
                else:
                    c_no = 0
                ax.scatter(x, y, c = ball_c[c_no][1], marker='o', s=200, zorder = 5)


        for ball in range(row):
            P_type = P_types[ball]
            df_P = df_L[df_L['球種'] == P_type]
            ax = fig.add_subplot(2,f'{row+1}',f'{ball+2+row}',title=f"{P_type} 対左")
            ax.title.set_size(15)
            #x軸設定
            ax.set_xlim(0, 160)
            ax.set_xticks([])
            x_v = [32,64,96,128]
            for i in range(len(x_v)):
                ax.axvline(x_v[i], ls = "-.", color = guide_c)
                if i == 0 or i == 3:
                    ax.axvline(x_v[i], color = strike_c,ymin=0.2, ymax=0.8)

            #y軸設定
            ax.set_yticks([])
            ax.set_ylim(200, 0)
            y_h = [40,80,120,160]
            for j in range(len(y_h)):
                ax.axhline(y_h[j], ls = "-.", color = guide_c)
                if j == 0 or j == 3:
                    ax.axhline(y_h[j], color = strike_c,xmin=0.2, xmax=0.8)

            #ストライクゾーン内の色設定
            r = patches.Rectangle( (32,40) , 96, 120, facecolor=strike_z_c,alpha=0.3)
            ax.add_patch(r)


            for ball2 in range(len(df_P)):
                x = df_P.iat[ball2,6]
                y = df_P.iat[ball2,7]
                co = df_P.iat[ball2,9]
                kind = df_P.iat[ball2,10]
                if kind == 1:
                    c_no = 3
                elif kind == 2:
                    c_no = 4
                elif kind == 3:
                    c_no = 5
                elif kind == 4:
                    c_no = 9
                elif kind == 5:
                    c_no = 10
                elif kind == 6:
                    if co != 0:
                        c_no = 7
                    else:
                        c_no = 6
                elif kind == 7:
                    c_no = 11
                elif kind == 8 or kind == 9 or kind == 10:
                    c_no = 12
                elif kind == 14:
                    c_no = 8
                elif kind == 0:
                    if co == 2:
                        c_no = 2
                    elif co == 1:
                        c_no = 1
                    else:
                        c_no = 0
                else:
                    c_no = 0
                ax.scatter(x, y, c = ball_c[c_no][1], marker='o', s=200, zorder = 5)

        os.makedirs(f"画像/{P_name}/投球チャート" , exist_ok=True)
        fig.savefig(f"画像/{P_name}/投球チャート/{P_name} {Day}.jpg")
        
        
        # OAuth認証 セッションを開始
        twitter = OAuth1Session(APIK, APIKS, AT, ATS)

        # 画像投稿
        files1 = {"media" : open(f"画像/{P_name}/球種割合/{P_name} {Day}.jpg", 'rb')}
        req_media1 = twitter.post(url_media, files = files1)

        # レスポンスを確認
        if req_media1.status_code != 200:
            print ("画像アップデート失敗: %s", req_media.text)
            exit()

        # Media ID を取得
        media_id1 = json.loads(req_media1.text)['media_id']
            
        # OAuth認証 セッションを開始
        twitter = OAuth1Session(APIK, APIKS, AT, ATS)

        # 画像投稿
        files2 = {"media" : open(f"画像/{P_name}/投球チャート/{P_name} {Day}.jpg", 'rb')}
        req_media2 = twitter.post(url_media, files = files2)

        # レスポンスを確認
        if req_media2.status_code != 200:
            print ("画像アップデート失敗: %s", req_media.text)
            exit()

        # Media ID を取得
        media_id2 = json.loads(req_media2.text)['media_id']

        # Media ID を付加してテキストを投稿
        params = {'status': f'{Day}日本シリーズ {P_name}\n球数:{len(_df)} 打者数:{Bat_c} 被安打:{sh+wh+th+Hh} 被本塁打:{Hh} 奪三振:{K} 四球:{fb} 死球:{db} 失点:{po}\n#プロ野球 #swallows',
                  "media_ids": f"{media_id1},{media_id2}"}
        req_media = twitter.post(url_text, params = params)

        # 再びレスポンスを確認
        if req_media.status_code != 200:
            print ("テキストアップデート失敗: %s", req_text.text)
            exit()

        print (f"{mem+1}人目OK")       
        
#seibulions #Seibulions_summary
```
